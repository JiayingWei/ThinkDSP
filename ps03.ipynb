**2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?**

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = (range(len(wave.ys)//2))
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
lags, corrs = autocorr(bitcoin)
thinkplot.plot(lags, corrs)
thinkplot.config(xlabel='lag (index)', ylabel='correlation', ylim=[-1, 1])
```

It does not appear that the auto correlation function decreases exponentially, it seems to actually decrease linearly which tells us that it has a **long-range dependance** which means that its current value depends quite abit on what its value was quite some time ago. There is no real indication of periodic behavior because its autocorrelation is also not periodic. I guess you can make an argument that if you extrapolate the results of the bitcoin prices it may be periodic, but that wasn't the question.

**4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.**

**You will find code in ps03starter.ipynb that will help with this.**

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(4, 12)    #different resolutions for the analysis
print(ns)
results1 = []
results2 = []
results3 = []
for N in ns:    #N is the number of frequencies that will be found
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result1 = %timeit -o dct.analyze1(ys, freqs, ts)
    result2 = %timeit -o dct.analyze2(ys, freqs, ts)
    result3 = %timeit -o scipy.fftpack.dct(ys)
    results1.append(result1)
    results2.append(result2)
    results3.append(result3)
```

```python
bests1 = [result1.best for result1 in results1]
bests2 = [result2.best for result2 in results2]
bests3 = [result3.best for result3 in results3]
```

```python
analyze1 = pyplot.plot(ns,bests1, label = 'analyze1', linewidth=2.0)
analyze2 = pyplot.plot(ns,bests2, label = 'analyze2',linewidth=2.0)
scipydct = pyplot.plot(ns,bests3, label = 'scipy.dct',linewidth=2.0)
pyplot.legend(loc=2)
thinkplot.config(title = "timeit results for wave analysis", xlabel = 'number of frequencies looked for', ylabel = 'amount of time (s)', xscale='log', yscale='log')
```

So looking at the timeit results for the different resolutions of the Gaussian noise, it appears that analyze2 (which takes advantage of the properties of symmetric matrixies) does better than analyze1 (which computes the inverses and such), which is expected due to the computation time of inverses vs transposes. However the scipy discrete cosine transform does the best, with its 2048 frequency resolution time  appearing to beat both analyze1 and analyze2 in computing their 16 resolution DCTs. So I'm guessing in terms of algorithemic complexity it goes analyze1 = most complex, analyze2 = less complex, scipy.dct = least complex. 

```python
analyze1_n = scipy.stats.linregress(numpy.log(ns),numpy.log(bests1))
analyze2_n = scipy.stats.linregress(numpy.log(ns),numpy.log(bests2))
scipydct_n =  scipy.stats.linregress(numpy.log(ns),numpy.log(bests3))
print(analyze1_n[0])
print(analyze2_n[0])
print(scipydct_n[0])
```

Looking at the slope of the lines in log scale allow us to figure out what power algorithem is because $log(o^n) = nlog(o)$ So analyze1 is $o^{2.2}$, analyze2 is $o^{1.8}$, scipydct is $o^{0.30}$.  

**5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:**<br />
**1.Break a long signal into segments.**<br />
**2.Compute the DCT of each segment.**<br />
**3.Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.**<br />
**To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.**<br />
**Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?**


```python
def inverse_dct_iv(amps):
    return dct.dct_iv(amps) * 2

def audio_compression(wave, amplitude):
    #breaks waves into segments
    segments = 1000
    new_amps = []
    ys = []
    ys_new = []
    duration = (wave.duration/segments)
    for i in range(segments):
        start = i*(wave.duration/segments)
        segment = wave.segment(start = start , duration = duration)
        #computes the DCT of each segments
        amps = dct.dct_iv(segment.ys)
        for amp in amps:
            if numpy.abs(amp) > amplitude:
                new_amps.append(amp)
        ys.append(inverse_dct_iv(new_amps))
        new_amps = []
        for j in range(len(ys[i])):
            ys_new.append(ys[i][j])
    ys = numpy.array(ys_new)
    new_wave = thinkdsp.Wave(ys,wave.framerate)
    return new_wave
    
```

```python
violin = thinkdsp.read_wave('92002__jcveliz__violin-origional.wav')
print(100-(len(violin.ys)/len(violin.ys))*100, '%')
violin.make_audio()
```

```python
violin_comp1 = audio_compression(violin, 2)
print(100 - (len(violin_comp1.ys)/len(violin.ys))*100, '%')
violin_comp1.make_audio()
```

```python
violin_comp2 = audio_compression(violin, 5)
print(100 - (len(violin_comp2.ys)/len(violin.ys))*100, '%')
violin_comp2.make_audio()
```

```python
violin_comp3 = audio_compression(violin, 100)
print(100 - (len(violin_comp3.ys)/len(violin.ys))*100, '%')
violin_comp3.make_audio()
```

```python
violin_comp4 = audio_compression(violin, 1000)
print(100 - (len(violin_comp4.ys)/len(violin.ys))*100, '%')
violin_comp4.make_audio()
```
