**1) At http://www.coindesk.com you can download the daily price of a BitCoin as a CSV file.  Read this file and compute the power spectrum of BitCoin prices as a function of time. Does it resemble white, pink, or Brownian noise?**

```python
with open('coindesk-2-1-2015.csv', 'rb') as bitcoin:
    reader = csv.reader(bitcoin)
    time = []
    value = []
    for row in reader:
        if len(row) > 1:
            time.append(row[0][11:18])
            value.append(row[1])
time = time[1::]
value = value[1::]

total_time = int(time[-1].replace(':','')) + 24000
total_time = str(total_time)
seconds = int(total_time[0:2]) * 3600 + int(total_time[2:3]) * 60 + int(total_time[4])
framerate = len(time)/seconds
bitcoin = thinkdsp.Wave(value, framerate)
bitcoin.plot()
thinkplot.config(title='Bit Coin Prices on 2-1-2015',xlabel='time (s)', ylabel='price ($)')
```

```python
bitcoin_spec = bitcoin.make_spectrum()
bitcoin_spec.plot_power()
thinkplot.config(title='Power density plot of Bitcoins on 2-1-2015',xlabel='frequency (Hz)', ylabel='power density', xscale='log', yscale='log', legend=False)
```

**2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?**

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = (range(len(wave.ys)//2))
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
lags, corrs = autocorr(bitcoin)
thinkplot.plot(lags, corrs)
thinkplot.config(xlabel='lag (index)', ylabel='correlation', ylim=[-1, 1])
```

It does not appear that the auto correlation function decreases exponentially, it seems to actually decrease linearly which tells us that it has a **long-range dependance** which means that its current value depends quite abit on what its value was quite some time ago. There is no real indication of periodic behavior because its autocorrelation is also not periodic. I guess you can make an argument that if you extrapolate the results of the bitcoin prices it may be periodic, but that wasn't the question.
