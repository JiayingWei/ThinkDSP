**1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.**

**The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?**

```python
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

def approx_equal(a, b, tol=1e-10):
    return sum(abs(a-b)) < tol

def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res
```

```python
N = 8
ts = numpy.arange(N) / N
freqs = numpy.arange(N)
freqs2 = freqs + N
args = numpy.outer(ts, freqs)
args2 = numpy.outer(ts,freqs2)
M = numpy.exp(i * PI2 * args)
M2 = numpy.exp(i * PI2 * args2)
amps = [1, 2, 3, 4, 5, 6, 7, 8]

print(approx_equal(M,M2))
```

If the IDFT matrixes are the same for frequency ranges (0 to N) and (N to 2N-1) that means that means that if you synthesize a wave from both those frequency ranges using the same amplitude you will end up with the same wave.

```python
ys1 = synthesize2(amps,freqs,ts)
ys2 = synthesize2(amps,freqs2,ts)

wave1 = thinkdsp.Wave(ys1, framerate = 8).plot()
```

```python
wave2 = thinkdsp.Wave(ys2, framerate = 8).plot()
```

**4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.**

**One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.**

```python
diff1 = thinkdsp.Wave(numpy.array([1, -1]), 2).make_spectrum().plot()
thinkplot.config(title = "filter of the window [-1, 1]")
```

So it makes sense that convolution with the window [1,-1] represents take the difference between each successive elements because it takes (X1 \* 1) + (X2 \* -1), which is basically X1 - X2. 

**If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.**

```python
diff2 = thinkdsp.Wave(numpy.array([-1/2, 2, -3/2]), 3).make_spectrum().plot()
thinkplot.config(title = "filter of the window [-1/2, 2, -3/2]", xlim = [0,1], ylim = [0,2])
```

So it looks like the second order difference approximation is still a linear upwards slope which makes still a high pass filter. However compared to the difference window it has a slightly steeper slope which means that it is a slightly more aggresive filter (not by much, aggresive is a very strong word for whats going on here).

**Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?**

**5) The gunshot recording I used in chap09.ipynb was clipped, which means that the loudest part gets cut off, which makes the rest of the sound seem louder than it should be, which exaggerates the reverb of the space.**

**But it looks like there is a better version of this recording at http://www.freesound.org/people/Kleeb/sounds/180960/**

```python
#Clipped gun shot
response1 = thinkdsp.read_wave('180961__kleeb__gunshots.wav')
response1 = response1.segment(start=0.26, duration=5.0)
response1.normalize()
#Un-clipped gunshot
response2 = thinkdsp.read_wave('180960__kleeb__gunshot.wav')
response2 = response2.segment(start=0.15, duration=5.0)
response2.normalize()
#Transfer function
transfer1 = response1.make_spectrum()
transfer2 = response2.make_spectrum()
```

```python
transfer2.plot()
thinkplot.config(title = 'un-clipped gunshot', xlabel = 'frequency (Hz)', ylabel = 'A')
```

So looking at the two transfer functions, one of the clipped, and the other of the un-clipped gunshot, you notise that the clipped one has alot more higher frequency components, this makes sense because clipped things have sharper edges which results in higher frequencies in the DFT.

**6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.**

**Hint: this works best if the duration of the wave is an integer number of periods.**

```python
parasig = thinkdsp.ParabolicSignal()
parasig.plot()
thinkplot.config(title = 'Parabolic waveform', xlabel = 'time(s)', ylabel = 'A')
```

```python
parasigspec = parasig.make_wave().make_spectrum()
parasigspec.plot()
thinkplot.config(title = 'Parabolic spectrum', xlabel = 'frequency(Hz)', ylabel = 'A')
```

```python
parasigspec.differentiate()
parasigspec.plot()
thinkplot.config(title = 'Parabolic differentiated spectrum', xlabel = 'frequency(Hz)', ylabel = 'A')
```
