**1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.**

**The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?**

```python
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

def approx_equal(a, b, tol=1e-10):
    return sum(abs(a-b)) < tol

def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res
```

```python
N = 8
ts = numpy.arange(N) / N
freqs = numpy.arange(N)
freqs2 = freqs + N
args = numpy.outer(ts, freqs)
args2 = numpy.outer(ts,freqs2)
M = numpy.exp(i * PI2 * args)
M2 = numpy.exp(i * PI2 * args2)
amps = [1, 2, 3, 4, 5, 6, 7, 8]

print(approx_equal(M,M2))
```

If the IDFT matrixes are the same for frequency ranges (0 to N) and (N to 2N-1) that means that means that if you synthesize a wave from both those frequency ranges using the same amplitude you will end up with the same wave.

```python
ys1 = synthesize2(amps,freqs,ts)
ys2 = synthesize2(amps,freqs2,ts)

wave1 = thinkdsp.Wave(ys1, framerate = 8).plot()
```

```python
wave2 = thinkdsp.Wave(ys2, framerate = 8).plot()
```

**2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).**

**Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.**

```python
N = 30
freqs = numpy.arange(N)
ts = numpy.arange(N)/N
fc = N/2
framerate = N
yslowpass = numpy.zeros(N)
yslowpass[:fc] = 1
yshighpass = numpy.zeros(N)
yshighpass[fc:] = 1

thinkplot.plot(freqs,yslowpass)
thinkplot.config(title = 'approximation of a lowpass window', xlabel = 'freq (Hz)', ylabel = 'A', ylim = [-.1,1.1])
```

```python
thinkplot.plot(freqs,yshighpass)
thinkplot.config(title = 'approximation of a highpass window', xlabel = 'freq (Hz)', ylabel = 'A', ylim = [-.1,1.1])
```

```python
lowpass = synthesize2(yslowpass,freqs,ts)
lowpasswave = thinkdsp.Wave(lowpass, framerate).plot()
thinkplot.config(title = 'lowpass window', xlabel = 'time', ylabel = 'A', ylim=[-6,6])
```

```python
highpass = synthesize2(yshighpass,freqs,ts)
highpasswave = thinkdsp.Wave(highpass, framerate).plot()
thinkplot.config(title = 'highpass window', xlabel = 'time', ylabel = 'A',ylim=[-6,6])
```

These windows were created from a relatively low resolution filter, it was purposely done so that I could observe the general trend of the window without needing to zooom in. It appears that both lowpass and highpass filters start off with a large spike at the beginning, but the lowpass window seems to have smaller upside down triangle waves, while the highpass window has rightside up triangle waves.
