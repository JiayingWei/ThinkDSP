```python
class ComplexSinusoid(thinkdsp.Sinusoid):
    """Represents a complex exponential signal."""
    
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        print(ts)
        i = complex(0, 1)
        phases = PI2 * self.freq * ts + self.offset
        print(phases)
        ys = self.amp * numpy.exp(i * phases)
        return ys
    
def synthesize1(amps, freqs, ts):
    components = [thinkdsp.ComplexSinusoid(freq, amp)
    for amp, freq in zip(amps, freqs)]
    signal = thinkdsp.SumSignal(*components)
    ys = signal.evaluate(ts)
    return ys
```

**Can you construct a vector of complex amplitudes that shifts all components by the same amount of time?**

The time it takes for a 100Hz signal to travel $1rad$ is $\frac{2\pi}{100}$ seconds. <br/>
The time it takes for a 200Hz signal to travel $1rad$ is $\frac{2\pi}{200}$ seconds. <br/><br/>
If you think about it, if you give them the same $\phi_0$ (or phase shift) it's actually giving the 100Hz signal a better headstart because it is slower. <br/><br/>
**_If you want everything to be shifted the same in the time domain, you need to make sure that the amount of advantage given to all the signals is the same._**<br/><br/>
In my example, you would want the 200Hz signal to be given twice the headstart the 100Hz signal is given because it is twice as fast. 


```python
def evenThePlayingField(phi_0, freqs):
    phi_0s = []
    for freq in freqs:
        phi_0s.append(freq/freqs[0] * 1j)
    return phi_0s
```

**In mathematics, the discrete Fourier transform (DFT) converts a finite list of equally spaced samples of a function into the list of coefficients of a finite combination of complex sinusoids, ordered by their frequencies, that has those same sample values.** <br/>

Translation: It takes in the ys, and then finds the amps of each signal

![wikipdiaDFT](wikipediaDFT.jpg)


```python
def wikipediaDFT(ys, n, framerate):
    N = len(ys)
    ts = numpy.arange(N)/N
    freqs = numpy.arange(N)*framerate/N  
    amps = []
    Xk = []
    for i in range(len(freqs)):
        Xk = ys[i] * numpy.exp(-1j * PI2 * freqs[i] * ts/N)
        amps.append(sum(Xk))
        Xk = []
    return amps
```

**3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.**

**Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.**


```python
#written with help from https://jakevdp.github.io/blog/2013/08/28/understanding-the-fft/

def wikipediaFFT(ys):
    if len(ys) % 2 == 0:
        pass
    elif len(ys) != 1:
        ys = ys[:-1]
    N = len(ys)
    W = numpy.exp(-1j * PI2 * numpy.arange(N) / N)
    if N == 1: 
        return ys
    else:
        evens = wikipediaFFT(ys[::2])
        odds = wikipediaFFT(ys[1::2])
        x0 = evens + W[:N/2] * odds
        x1 = evens + W[N/2:] * odds
        return numpy.concatenate([x0,x1])
```

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
```

```python
ns = 2 ** numpy.arange(4, 12)    #different resolutions for the analysis
print(ns)
results1 = []
results2 = []
results3 = []
for N in ns:    #N is the number of frequencies that will be found
    ys = noise.ys[:N]
    result1 = %timeit -o autocorr(thinkdsp.Wave(ys, noise.framerate))
    result2 = %timeit -o fft_autocorr(ys)
    result3 = %timeit -o numpy.correlate(ys,ys,'same')
    results1.append(result1)
    results2.append(result2)
    results3.append(result3)
```

```python
bests1 = [result1.best for result1 in results1]
bests2 = [result2.best for result2 in results2]
bests3 = [result3.best for result3 in results3]
```

```python
analyze1 = pyplot.plot(ns,bests1, label = 'Allen autocorr', linewidth=2.0)
analyze2 = pyplot.plot(ns,bests2, label = 'Allen FFT autocorr',linewidth=2.0)
scipydct = pyplot.plot(ns,bests3, label = 'numpy autocorr',linewidth=2.0)
pyplot.legend(loc=2)
thinkplot.config(title = "timeit results for autocorrelation", xlabel = 'length of input file', ylabel = 'amount of time (s)', xscale='log', yscale='log')
```
