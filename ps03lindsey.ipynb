Problem 1 - **Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?**

```python
ys = []
file_name = 'coindesk-2-1-2015.csv'
# file_name = 'coindesk-2-1-2015.csv'
with open(file_name, 'rb') as csvfile:
    bit_read = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in bit_read:
        if len(row) > 1:
            data = row[1]
            ys.append(data)
ys = np.array(ys[1::])
ts = np.linspace(0, len(ys), len(ys))
bit_wave = thinkdsp.Wave(ys, 1)
bit_wave.plot()
thinkplot.config(xlabel='time since opening (s)', ylabel='price (USD)', legend=False)
```

```python
bit_spec = bit_wave.make_spectrum()
bit_spec.plot()
thinkplot.show(xlabel='frequency (Hz)', ylabel='power density', xscale='log', yscale='log')
```

Problem 2- **Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?**

```python
def corrcoef(xs, ys):
    return np.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

lags, cors = autocorr(bit_wave)
thinkplot.Plot(lags, cors)
thinkplot.show(xlabel='lag', ylabel='correlation')
```

**Comments: ** The autocorrelation of bitcoin prices seems to decay fairly linearly actually and then starts to come back up. I am not really sure from this plot wheter or not the correlation is periodic because althought it looks like it is starting to come back up at the same rate that the correlation decreased, there is not yet enough data to suggest that this trend would continue and that if it did it would repeat. It is interesting to note that as the lag increases there does seem to be an equal amount/ magnitude of positive and negative correlation, which may hint at a periodic behavior.

Problem 3 - **Use autocorrelation to estimate the pitch of the saxophone recording in 100475__iluppai__saxophone-weep.wav.  What happens if you filter out the fundamental tone?  Is the estimate based on the autocorrelation function consistent with your perceived pitch?**

Autocorrelation Estimate with the fundamental tone still present

```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
segment = wave.segment(start=0.4, duration=0.03)
# wave.plot()
segment.plot()
```

```python
spec = wave.make_spectrum()
spec.plot()
```

```python
gram = wave.make_spectrogram(seg_length=1004)
gram.plot()
thinkplot.config(ylim=[0, 6000])
```

```python
lags, corrs = autocorr(segment)
thinkplot.plot(lags, corrs)
thinkplot.config(xlabel='lag (index)', ylabel='correlation', ylim=[-1, 1])
```

```python
low, high = 140, 160
lag = np.array(corrs[low:high]).argmax() + low
period = lag / float(wave.framerate)
frequency = 1/float(period)
low_guess = wave.framerate / float(lag-1)
high_guess = wave.framerate / float(lag+1)
print(high_guess, low_guess, frequency)
```
